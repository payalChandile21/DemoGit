# <span style="color:blue">Python if...else Statement
In computer programming, the if statement is a conditional statement. It is used to execute a block of code only when a specific condition is met. For example,

Suppose we need to assign different grades to students based on their scores.

- If a student scores above 90, assign grade A
- If a student scores above 75, assign grade B
- If a student scores above 65, assign grade C
- These conditional tasks can be achieved using the if statement.

## <span style="color:blue"> Python if Statement
An if statement executes a block of code only if the specified condition is true.

#### Syntax

Here, if the condition of the if statement is:

- True - the body of the if statement executes.
- False - the body of the if statement is skipped from execution.

## <span style="color:blue"> Let's try a few examples with if condition and explore comparison operators.

<span style="color:red">Demonstrating the use of less than ' < ' operator 

```python
#using less than ('<')
number = 10

# Check if number is less than 20
if number < 20:
    print('Number is less than 20')

print('This statement always executes')
```

<span style="color:red"> Demonstrating the use of greater than equal '>=' operator 

```python
#Greater than equal to ('>=')
number = 10

# Check if number is greater than or equal to 10
if number >= 10:
    print('Number is 10 or greater')

print('This statement always executes')
```

<span style="color:red"> Demonstrating the use of less than equal '<=' operator 

```python
#Less than or equal to ('<=')
number = 10

# Check if number is less than or equal to 10
if number <= 10:
    print('Number is 10 or less')

print('This statement always executes')
```

<span style="color:red"> Demonstrating the use of equal to '==' operator 

```python
#equal to ('==')
number = 10

# Check if number is equal to 10
if number == 10:
    print('Number is exactly 10')

print('This statement always executes')
```

<span style="color:red"> Demonstrating the use of not equal to '!=' operator 

```python
#not equal to ('!=')
number = 10

# Check if number is not equal to 5
if number != 5:
    print('Number is not 5')

print('This statement always executes')
```

## <span style="color:blue"> Python if...else Statement
An if statement can have an optional else clause. The else statement executes if the condition in the if statement evaluates to False.

## <span style="color:blue"> Let's try a few examples with membership in list .

<span style="color:red">  Demonstrating if-else

```python
#Equality (==)
number = 10

# Check if number is equal to 10
if number == 10:
    print('Number is exactly 10')
else:
    print('Number is not 10')

print('This statement always executes')
```

```python
#Inequality (!=)
number = 10

# Check if number is not equal to 5
if number != 5:
    print('Number is not 5')
else:
    print('Number is 5')

print('This statement always executes')
```

#### <span style="color:red"> 1. Write a python program to check if the number is present in the list. 

```python
#Checking Membership in a List ('in')
number = 10
numbers_list = [1, 2, 10, 20]

# Check if number is in the list
if number in numbers_list:
    print('Number is in the list')

print('This statement always executes')

```

#### <span style="color:red">2. Write a python program to check if a string contains, starts with, or ends with a substring

```python
# Get a string and a substring from the user
main_string = input("Enter the main string: ")
substring = input("Enter the substring: ")

# Check if the substring is in the main string, starts with, or ends with it

if main_string.startswith(substring):
    print(f"The main string starts with '{substring}'.")
elif main_string.endswith(substring):
    print(f"The main string ends with '{substring}'.")
elif substring in main_string:
    print(f"The substring '{substring}' is found in the main string.")
else:
    print(f"The substring '{substring}' is not found in the main string.")
```

## <span style="color:blue">Logical Operators (AND, OR, NOT)

1. The And operator is used to verify where both conditions associated with it are True Simultaneously. It is represented as "X and Y".
2. The Or Operator is used to verify that either of the associated conditions is true. It is represented as "X or Y".
3. The Not Operator is associated with a single condition. It inverts the results, i.e., true is changed to false, and false is changed to true. It is represented as "not X".

## <span style="color:blue"> NESTED if...else

## <span style="color:blue">Let's try a few examples

<span style="color:red"> Write a Python program that checks if a given number is positive, negative, or zero using nested if statements. Print an appropriate message based on the value of the number.

```python
number = 5

# outer if statement
if number >= 0:
    # inner if statement
    if number == 0:
      print('Number is 0')
    
    # inner else statement
    else:
        print('Number is positive')

# outer else statement
else:
    print('Number is negative')
```

<span style="color:red"> Write a Python program that uses nested if-else statements to check if a given number is positive, negative, or zero. Print an appropriate message based on the value of the number, and include a statement that always executes regardless of the number's value.

```python
#Greater Than (>)
number = 10

# Nested if-else for checking if number is greater than 0
if number > 0:
    print('Number is positive')
else:
    if number == 0:
        print('Number is zero')
    else:
        print('Number is negative')

print('This statement always executes')
```
