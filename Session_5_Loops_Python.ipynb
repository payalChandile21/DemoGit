## <span style="color:blue"> for Loop with Python range()
In Python, the range() function returns a sequence of numbers. For example,

Here, range(4) returns a sequence of 0, 1, 2 ,and 3.

Since the range() function returns a sequence of numbers, we can iterate over it using a for loop. For example,

```python
# iterate from i = 0 to i = 3
for i in range(4):
    print(i)
```

Here, we used the for loop to iterate over a range from 0 to 3.

This is how the above program works.

<span style="color:red">Write a Python program that uses a for loop to print numbers from 0 to n-1, where n is a given integer.

```python
# Example 1: Loop from 0 to n-1
n = 5
for number in range(n):
    print(number, end=' ')  # Output: 0 1 2 3 4
print()
```

## <span style="color:blue">for loop with (start, stop, step)
The range() function in Python can take three arguments: start, stop, and step. This allows you to create a sequence of numbers with a custom step.

<span style="color:red">Write a Python program that demonstrates the use of the range() function with positive and negative steps.

```python
# Positive step
for num in range(0, 10, 2):
    print(num, end=' ')  # Output: 0 2 4 6 8
print()

# Negative step
for num in range(10, 0, -2):
    print(num, end=' ')  # Output: 10 8 6 4 2
print()
```

### <span style="color:blue">The break Statement
With the break statement we can stop the loop before it has looped through all the items:


```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

### <span style="color:blue">The continue Statement
With the continue statement we can stop the current iteration of the loop, and continue with the next:


```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

### <span style="color:blue">Nested Loops
A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop":

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

## <span style="color:blue">while loop with conditions
In Python, we use the while loop to repeat a block of code until a certain condition is met. For example,


```python
number = 1

while number <= 3:
    print(number)
    number = number + 1
```

```python
i = 0
while i < 5:
    if i == 3:
        break
    print(i)
    i += 1
```

```python
i = 0
while i < 5:
    if i == 3:
        i += 1
        continue
    print(i)
    i += 1
```

In the above example, we have used a while loop to print the numbers from 1 to 3. The loop runs as long as the condition number <= 3 is satisfied.

### <span style="color:blue">while Loop Syntax



1. The while loop evaluates the condition.
2. If the condition is true, body of while loop is executed.The condition is evaluated again.
3. This process continues until the condition is False.
4. Once the condition evaluates to False, the loop terminates.

#### <span style="color:red">1.Write a Python program using a while True loop to create a menu-driven calculator. The program should continuously display a menu with options to add, subtract, multiply, divide, or exit. Implement each operation using user input for numbers, ensuring proper handling of arithmetic operations and division by zero.

```python
while True:
    print("Menu:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        result = num1 + num2
        print(f"Result: {num1} + {num2} = {result}")

    elif choice == '2':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        result = num1 - num2
        print(f"Result: {num1} - {num2} = {result}")

    elif choice == '3':
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        result = num1 * num2
        print(f"Result: {num1} * {num2} = {result}")

    elif choice == '4':
        num1 = float(input("Enter dividend: "))
        num2 = float(input("Enter divisor: "))
        if num2 != 0:
            result = num1 / num2
            print(f"Result: {num1} / {num2} = {result}")
        else:
            print("Error: Division by zero!")

    elif choice == '5':
        print("Exiting the program. Goodbye!")
        break  # Exit the loop and end the program

    else:
        print("Invalid choice. Please enter a number from 1 to 5.")
```
