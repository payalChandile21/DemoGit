# <span style="color:blue">Python Variables
A variable in Python is a name that is used to refer to memory location. It is a way to store data for later use in your program. Variables are essential in programming as they allow you to store, retrieve, and manipulate data.

# <span style="color:blue">Python Datatypes
Python provides a variety of data types to handle different kinds of data. Understanding these data types is essential for effective programming in Python. Here's an overview of the main data types in Python:

1. Numeric Types:
int: Represents whole numbers without a fractional part. Examples include 5, -10, and 0.
float: Represents numbers with a fractional part. Examples include 3.14, -0.001.
complex: Represents complex numbers with real and imaginary parts. Examples include 1 + 2j.

2. str (String): A sequence of characters enclosed in single, double, or triple quotes. Examples include "hello", 'Python'.

3. Boolean Type:
bool: Represents one of two values: True or False. 

4. list: An ordered collection of items, which can be of different types. Lists are mutable, meaning their content can be changed. Examples include [1, 2, 3], ["apple", "banana"].

5. dict (Dictionary): A collection of key-value pairs. Dictionaries are unordered and mutable. Examples include {"name": "Alice", "age": 25}.

## <span style="color:blue"> Type Conversion/ Type Casting
You can convert from one type to another with the int(), float(), and complex() methods

```python
# Initial values
x = 5       # int
y = 3.7     # float
z = 5 + 3j  # complex

# Convert from int to float
a = float(x)

# Convert from float to int
b = int(y)

# Convert from complex to int (taking the real part only)
c = int(z.real)

# Convert from int to string
d = str(x)

# Convert from float to string
e = str(y)

# Convert from complex to string
f = str(z)

# Print the converted values
print(a)  # Output: 5.0
print(b)  # Output: 3
print(c)  # Output: 5
print(d)  # Output: '5'
print(e)  # Output: '3.7'
print(f)  # Output: '(5+3j)'

# Print the types of the converted values
print(type(a))  # Output: <class 'float'>
print(type(b))  # Output: <class 'int'>
print(type(c))  # Output: <class 'int'>
print(type(d))  # Output: <class 'str'>
print(type(e))  # Output: <class 'str'>
print(type(f))  # Output: <class 'str'>
```

```python
x = float(1)     # x will be 1.0
y = float(2.8)   # y will be 2.8
z = float("3")   # z will be 3.0
w = float("4.2") # w will be 4.2
```

# <span style="color:blue">Python Comments
Comments can be used to explain Python code.

Comments can be used to make the code more readable.

Comments can be used to prevent execution when testing code.

<span style="color:red">Comments starts with a #, and Python will ignore them

```python
#This is a comment
print("Hello, World!")
```
