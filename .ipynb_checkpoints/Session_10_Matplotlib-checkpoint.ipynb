### <span style="color:blue">What is Matplotlib?
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib.pyplot as plt
```

### <span style="color:blue">Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## <span style="color:blue">Multiple Points

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### <span style="color:blue">Plot with Marker
Markers
You can use the keyword argument marker to emphasize each point with a specified marker:

```python
# Plot with Marker
years = [2021, 2020, 2019, 2018]
prices = [79990, 20000, 55000, 22000]

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

plt.plot(years, prices, marker='o',ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.title('Prices Over Years', fontdict=font)
plt.xlabel('Year', fontdict=font)
plt.ylabel('Price ($)', fontdict=font)
plt.grid(True)
plt.show()
```

### <span style="color:blue">Linestyle
You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line

```python
# Plot with Linestyle
plt.plot(years, prices, linestyle='--')
plt.title('Prices Over Years')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

### <span style="color:blue">Subplot

```python
# Subplot
plt.subplot(2, 1, 1)
plt.plot(years, prices)
plt.title('Prices Over Years')

plt.subplot(2, 1, 2)
horsepower = [670, 139, 450, 158]
plt.plot(years, horsepower)
plt.title('Horsepower Over Years')

plt.tight_layout()
plt.show()
```

### <span style="color:blue">Scatter Plot

```python
# Scatter Plot
plt.scatter(years, prices)
plt.title('Prices Over Years')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

### <span style="color:blue">Bar Plot

```python
# Bar Plot
plt.bar(years, prices, color='blue', width=0.4)
plt.title('Prices Over Years')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.show()
```

### <span style="color:blue">Horizontal Bar Plot


```python
# Horizontal Bar Plot
plt.barh(years, prices, color='green')
plt.title('Prices Over Years')
plt.xlabel('Price ($)')
plt.ylabel('Year')
plt.show()
```

### <span style="color:blue">Histogram

```python
# Histogram
plt.hist(prices, bins=5, color='purple')
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

### <span style="color:blue">Pie Chart

```python
# Pie Chart

sizes = [79990, 20000, 55000, 22000]
plt.pie(sizes)
plt.show()
```

```python
# Pie Chart
labels = 'Tesla', 'Toyota', 'Ford', 'Honda'
sizes = [79990, 20000, 55000, 22000]
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Price Distribution by Manufacturer')
plt.show()
```
