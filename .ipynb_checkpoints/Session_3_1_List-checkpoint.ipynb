## <span style="color:blue"> Slicing of a List in Python
In Python, it is possible to access a section of items from the list using the slicing operator :. For example,

<span style="color:red">Write a Python program to:

- <span style="color:red">Print the items from index 2 to index 4 in my_list.
- <span style="color:red">Print the items from index 5 to the end of my_list.
- <span style="color:red">Print all items in my_list from the beginning to the end.

```python
my_list = ['p', 'r', 'o', 'g', 'r', 'a', 'm']

# items from index 2 to index 4
print(my_list[2:5])

# items from index 5 to end
print(my_list[5:])

# items beginning to end
print(my_list[:])
```

## <span style="color:blue">Add Elements to a List From Other Iterables
We use the extend() method to add elements to a list from other iterables. For example,

<span style="color:red">Write a Python program that creates a list numbers with elements [1, 3, 5], then extends it by adding elements from another list even_numbers with elements [2, 4, 6]. Print the updated numbers list.

```python
numbers = [1, 3, 5]
print('Numbers:', numbers)

even_numbers  = [2, 4, 6]

# adding elements of one list to another
numbers.extend(even_numbers)

print('Updated Numbers:', numbers) 
```

## <span style="color:blue"> Remove One or More Elements of a List
The del statement removes one or more items from a list. For example,



<span style="color:red">Write a Python program that initializes a list names with elements ['John', 'Eva', 'Laura', 'Nick', 'Jack'], and performs the following deletions:

- <span style="color:red">Deletes the second item in the list using del.
- <span style="color:red">Deletes items from index 1 to index 3 using del

```python
names = ['John', 'Eva', 'Laura', 'Nick', 'Jack']

# deleting the second item
del names[1]
print(names)

# deleting items from index 1 to index 3 
del names[1: 4]
print(names) # Error! List doesn't exist.
```

## <span style=color:blue>Reverse Order
What if you want to reverse the order of a list, regardless of the alphabet?

The reverse() method reverses the current sorting order of the elements.

<span style="color:red">Write a Python program that initializes a list thislist with elements ["Java", "Python", "Ruby", "html", "css", "GoLang"], reverses the list using the reverse() method, and then prints the reversed list.

## <span style="color:blue">Copy a List
You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.


<span style ="color:red">Write a Python program that initializes a list thislist with elements ["Java", "Python", "Ruby", "html", "css", "GoLang"], creates a new list mylist by copying thislist using the list() constructor, and then prints mylist.

```python
thislist = ["Java", "Python", "Ruby", "html","css","GoLang"]
mylist = list(thislist)
print(mylist)
```
